<h1>Census 2011(India)</h1>
<h3>Data Analysis</h3>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("censusdata.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.head()

```

<h2>Religion based composition of India</h2>

```python
dataset=(sum(df['Hindus']),sum(df['Muslims']),sum(df['Christians']),sum(df['Sikhs']),sum(df['Buddhists']),sum(df['Jains']))
palette_color = seaborn.color_palette('bright')
plt.pie(dataset, labels=['Hindus','Muslims','Christians','Sikhs','Buddhists','jains'], colors=palette_color, autopct='%.0f%%')
plt.legend()
plt.show()
```

<h2>Statewise Population of the india</h2>

```python
from collections import Counter 

sns.barplot(x = 'State_name',
            y = 'Population',
            data = df)

plt.xticks(rotation = 45, ha = 'right')
 

plt.show()
```

<h2>Show the Male vs Female ratio of Nagaland</h2>

```python
newdf=df[df['State_name'].isin(['NAGALAND'])]
dataset=(sum(newdf['Male']),sum(newdf['Female']))
palette_color = seaborn.color_palette('bright')
plt.pie(dataset, labels=['Male','Female'], colors=palette_color, autopct='%.0f%%')
plt.legend()
plt.show()
```

<h2>Find out the literacy rate of Rajasthan</h2>

```python
newdf1=df[df['State_name']=='RAJASTHAN']
total_population=sum(newdf1['Population'])
total_literate=sum(newdf1['Literate'])
print((total_literate/total_population)*100)

```
